```python
df = DataExtract('AAPL')
df_vero = df.reed()
```

```python
# Supponiamo che tu abbia già un DataFrame df con i tuoi dati

# Crea il modello di regressione lineare con la colonna 'Adj Close' e una sequenza di 30 giorni
model = LinearRegressionModel(df_vero, seq_length=30)

# Suddividi i dati in set di addestramento e di test
X_train, X_test, y_train, y_test = model.split()

# Allena il modello
model.fit(X_train, y_train)

# Previsioni sul set di test
y_pred = model.predict(X_test)

# Calcola l'errore MSE
mse = model.evaluate(X_test, y_test)
print(f"Mean Squared Error: {mse:.4f}")

# Prevedi il valore del prossimo giorno (ultimo valore nel test)
next_day_pred = model.predict(X_test[-1:].reshape(1, -1))
print(next_day_pred)
# print(f"Previsione del prezzo per il prossimo giorno: {next_day_pred[0]:.2f}")
```
